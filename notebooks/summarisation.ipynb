```python
import json
import seaborn as sns
from os.path import exists
import pandas as pd
```

```python
from matplotlib import rcParams

# figure size in inches
rcParams['figure.figsize'] = 20,12
```

```python
MODELS = [
    "text-davinci-002",
    "text-curie-001",
    "text-babbage-001",
    "bigscience/bloom",
    "EleutherAI/gpt-j-6B",
    "gpt-neox-20b",
    "facebook/opt-13b",
    "facebook/opt-30b",
    "co:here",
    "jurassic1-jumbo",
    "ul2",
    "google/flan-t5-xxl"
]

prompts = [
    "1",
    "2",
    "3",
    "4",
    "5",
    "bloom1",
    "bloom2",
    "bloom3",
    "bloom4",
    "bloom5"
]

n_shots = [0,1]
```

```python
from sklearn.metrics import precision_recall_fscore_support
import evaluate
```

```python
log_data = []
for model in MODELS:
    for prompt in prompts:
        for n_shot in n_shots:
            prompt_name = prompt
            model_name = model.replace('/', '-')
            log_file = f"../llm_logs/summarisation-{model_name}-{prompt_name}-{n_shot}.json"
            if exists(log_file):
                with open(log_file) as f:
                    data = json.load(f)
                
                preds = []
                labels = []
                for d in data['data']:
                    labels.append(d['label'])
                    pred = d['pred']
                    input_markers = ["Text: ", "Article: "]
                    for input_marker in input_markers:
                        if pred.find(input_marker) >= 0:
                            pred = pred[:pred.find(input_marker)]
                    preds.append(pred)
                rouge = evaluate.load("rouge")
                metrics = rouge.compute(predictions=preds, references=labels)

                #precision_recall_fscore_support(labels, preds, average="macro")
                log_data.append({
                    "model": model,
                    "prompt": prompt_name,
                    "n_shot": n_shot,
                    "rougeL": data['metrics']['rougeL'],
                    "rouge1": data['metrics']['rouge1'],
                })
                
                data['metrics'] = metrics
                with open(log_file, 'w') as f:
                    json.dump(data, f, indent=4)
            print("n_shot", n_shot)
        print("prompt", prompt)
    print("model", model)
```

```python
model_map = [
    ("text-davinci-002","GPT3-davinci"),
    ("text-curie-001","GPT3-curie"),
    ("text-babbage-001","GPT3-babbage"),
    ("bigscience/bloom","BLOOM"),
    ("EleutherAI/gpt-j-6B","GPT-J-6B"),
    ("gpt-neox-20b","GPT-NeoX-20B"),
    ("facebook/opt-13b","OPT-13B"),
    ("facebook/opt-30b","OPT-30B"),
    ("co:here","Co:here"),
    ("jurassic1-jumbo","Jurassic-1 Jumbo"),
    ("ul2","UL2"),
    ("google/flan-t5-xxl","Flan-T5-xxl")
]
for m in model_map:
    for l in log_data:
        if l['model'] == m[0]:
            l['model'] = m[1]
```

```python
df = pd.DataFrame(log_data)
```
